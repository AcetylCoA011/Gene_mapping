#  Rank the Go terms and the relevant genes
We input the ranked genes into Gorilla:http://cbl-gorilla.cs.technion.ac.il/, and obtained significant GO terms for positive and negative gene lists. 
Then, we assign the corresponding r values to each significant genes to check the correlation level between them with the MS regional progression.

```python
import pandas as pd
import re
```

```python
process = pd.read_excel('./positive/GO_positive.xlsx')
process_negative = pd.read_excel('./negative/GO_negative.xlsx')
```

```python
process_negative
```

```python
# get rid of GO terms which with p-value and FDR q-value larger than 0.05
pd_1_1 = process.sort_values(by=['Enrichment'], ascending=False)
valid_index = []
for k, (i, j) in enumerate(pd_1_1[['P-value', 'FDR q-value']].values):
    if i < 0.05 and j < 0.05:
       valid_index.append(k)
pd_1_2 = pd_1_1.iloc[valid_index]
pd_2_1 = process_negative.sort_values(by=['Enrichment'], ascending=False)
valid_index = []
for k, (i, j) in enumerate(pd_2_1[['P-value', 'FDR q-value']].values):
    if i < 0.05 and j < 0.05:
       valid_index.append(k)
pd_2_2 = pd_2_1.iloc[valid_index]
process_sort = pd_1_2
process_negative_sort = pd_2_2
```

```python
process_negative_sort
```

```python
# output files only contain significant GO terms
process_sort.to_excel('./positive/process_sort.xlsx',index = False)
process_negative_sort.to_excel('./negative/process_negative_sort.xlsx',index = False)
```

```python
#add r values to each gene

def func(s):
    s = s.strip('[').strip(']')
    pattern = re.compile(r'([A-Z\-0-9]+)  -  ([a-z\s0-9,\-\(\)]+)')
    s_list = pattern.findall(s)
    s_list = [[i, j.strip().strip(',')] for i,j in s_list]
    return s_list

def func_2(file_xlsx):
    target_file = file_xlsx
    dataframe = pd.read_excel(target_file)
    outcome_data = []
    #df.value transfers columns into array
    for line_data in dataframe.values:   
        head_columns = line_data[:9]
        target_column = line_data[9]
        s_list = func(target_column)
        for s_item in s_list:
            outcome_data.append(list(head_columns) + s_item)
    target_df = pd.DataFrame(outcome_data, columns=list(dataframe.columns)[:-1] + ['Gene']+['full name'])
    return target_df
 
```

```python
positive_mod = func_2('./positive/process_sort.xlsx')
negative_mod = func_2('./negative/process_negative_sort.xlsx')
```

```python
gene_r = pd.read_csv('../data/id_label/result_for_All/resultAll4.csv')
```

```python
gene_with_r = pd.DataFrame(gene_r, columns=['gene_symbol','r'])
```

```python
gene_with_r
```

```python
positive_r = positive_mod.merge(gene_with_r,left_on="Gene", right_on="gene_symbol").drop(labels='gene_symbol', axis=1).sort_values(by='Enrichment', ascending = False)
negative_r = negative_mod.merge(gene_with_r,left_on="Gene", right_on="gene_symbol").drop(labels='gene_symbol', axis=1).sort_values(by='Enrichment', ascending = False)
```

```python
negative_r = pd.merge(left=negative_mod, right=gene_with_r, how='left', left_on="Gene", right_on="gene_symbol")
```

```python
#genes in these files are now with their corresponding r values and GO terms. 
positive_r.to_excel('./positive/process_r.xlsx', index = False)
negative_r.to_excel('./negative/process_negative_r.xlsx', index = False)
```
