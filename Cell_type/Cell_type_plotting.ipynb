# Cell type correlation

We aimed to combine the genetic expression data with the cell fraction data to investigate the correlation between the cell types with the temporal progression of brain atrophic regions in MS

```python
import pandas as pd
```

```python
# read merge_data (ID + EBM stage) for all MS patients
merged_data_All = pd.read_csv('../data/id_label/result_for_All/merged_data_All.csv')
```

```python
merged_data_All.drop(['Unnamed: 0'], axis = 1)
```

```python
wanted_EBM = pd.DataFrame(merged_data_All, columns=['well_id','EBM_Stage'])
```

```python
wanted_EBM
```

```python
cell_type = pd.read_csv('cell_type.csv')
```

```python
wanted_cell = cell_type.drop(['Unknown'],axis=1)
```

```python
wanted_cell = wanted_cell.rename(columns={'Unnamed: 0': 'well_id'}, inplace=False)
```

```python
wanted_cell
```

```python
merged_cell_type = wanted_EBM.merge(wanted_cell,left_on="well_id", right_on="well_id")
```

```python
merged_cell_type
```

```python
merged_cell_type.columns[2:8]
```

```python
# run the correlation analysis
from scipy.stats import spearmanr
result = []
for cell in merged_cell_type.iloc[:,2:8]:
   r, p = spearmanr(merged_cell_type.EBM_Stage, merged_cell_type[cell])
   result.append([r,p])
cell_cor = pd.DataFrame(result, columns=["r","p"],
index=merged_cell_type.columns[2:8].T)
# cell_cor
cell_cor.sort_values(by='r', ascending = True)
```
